```python
import requests
import json
import glob
import time
import numpy as np
 
def microservice_image(website, keyrwds):
    
    url = "http://127.0.0.1:8082/scape"  
    
    form_data = {
    'website':  website,
    'keywords':  json.dumps(keyrwds),
    }

    # content_type = 'application/octet-stream'

    response = requests.post(url, data=form_data)
    print(response)
    meta_data = json.loads(response.content)
    print(meta_data)
    return meta_data
```

```python
wald_key = {"Process control solution": ["Process improvement","process control"], 
            "Audit trail": ["Traceability","Time stamp"], "User management": [
                "User governance","User roles","Logging","Login credentials"], 
            "Real-time processing": ["Real-time control","Real-time improvement"], 
            "Electronic signatures": ["digital signature"], "Validation": ["Certificate"], 
            "Inspection plan management": ["Measurement Plan"], "Disaster recovery": ["Results Recovery"], 
            "Automatic archiving": ["Archiving"], "Cloud archiving": ["Central storage"], 
            "Document management": [], "21 CFR part 11": [], "21 CFR part 820": [], "ISO 13485": [], 
            "EN ISO 15883": [], "ISO 14971": [], "GxP": [], "GMP": [], "Regulatory Authority requirements": [], 
            "Data management solution": []}
```

```python
websites = ['https://www.boehme-weihs.de/en/',  
            'https://agosense.com/de/produkte/agosensefidelia/',  
            'https://www.bosch-connected-industry.com/en/',  
            'https://www.capture3d.com/', 
            'https://www.creaform3d.com/', 
            'https://www.diondo.com/', 'https://www.faro.com/', 
            'https://www.foodlogiq.com/', 
            'https://www.hexagonmi.com/products/', 
            'https://www.innovalia-metrology.com/', 
            'https://www.innovmetric.com/en/products/', 'https://www.inos-automation.de/', 
            'https://www.keyence.com/', 'https://kotem.com/']
res = {}
res = json.load(open('waldamar_websites_scrape.json', 'r'))
```

```python
print(list(res.keys()))
np.random.shuffle(websites)
#websites = ['https://www.hexagonmi.com/products/']
for website in websites:

    if website not in res: # res[website]['num_pages_crawled'] < 2:
        t = time.time()
        res[website] = microservice_image(website, wald_key)
        print(website)
        print(time.time() -t)
        json.dump(res, open('waldamar_websites_scrape.json', 'w'))
    
```

```python

for url in res:
    print (f"-------------")
    print (f"website: {url}")
    print (f"-------------")
    print(f"pages scraped: {res[url]['num_pages_crawled']}")
    
    r_ = res[url]['results']
    for kywrd in r_:
        print(f"keyword: {kywrd}\n score: {r_[kywrd]['score']} {r_[kywrd]['url']} ")
    print("\n")
    if 'errors_on' in res[url]:
        errs_ = res[url]['errors_on']
        for pgs in errs_:
            print(f"error on {pgs}: {errs_[pgs]}")
```
